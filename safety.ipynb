```python
labels = pd.read_csv('labels.csv')
```

## Condense to single data point

Get max of each factor for each ride

```python
# Create empty condensed DF
condensed = pd.DataFrame(0.0, index=np.arange(len(labels)), columns=['bookingID', 'acceleration_x', 'acceleration_y', 'acceleration_z', 'gyro_x', 'gyro_y', 'gyro_z', 'Label'])
```

```python
for i in range(len(labels)):
    condensed.iloc[i,0] = labels.iloc[i,0]
    condensed.iloc[i,7] = labels.iloc[i,1]
    df = data_dropped[data_dropped['bookingID']==condensed.iloc[i,0]]
    condensed.iloc[i,1] = max(abs(df['acceleration_x']))
    condensed.iloc[i,2] = max(abs(df['acceleration_y']))
    condensed.iloc[i,3] = max(abs(df['acceleration_z']))
    condensed.iloc[i,4] = max(abs(df['gyro_x']))
    condensed.iloc[i,5] = max(abs(df['gyro_y']))
    condensed.iloc[i,6] = max(abs(df['gyro_z']))
```

```python
condensed
```

## Simple classification 

```python
# Simple logistic regression
model = LogisticRegression()
model.fit(condensed[['acceleration_x', 'acceleration_y', 'acceleration_z', 'gyro_x', 'gyro_y', 'gyro_z']], condensed['Label'])
model.score(condensed[['acceleration_x', 'acceleration_y', 'acceleration_z', 'gyro_x', 'gyro_y', 'gyro_z']], condensed['Label'])
```

```python
# Simple random forest classifier
RF = RandomForestClassifier(n_estimators=500, min_samples_leaf=1000, oob_score=True, random_state=2, verbose=1)
RF.fit(condensed[['acceleration_x', 'acceleration_y', 'acceleration_z', 'gyro_x', 'gyro_y', 'gyro_z']], condensed['Label'])
RF.score(condensed[['acceleration_x', 'acceleration_y', 'acceleration_z', 'gyro_x', 'gyro_y', 'gyro_z']], condensed['Label'])
```

## Distribution of each factor

```python
labels['label'].value_counts()
```

```python
# Get nth percentile value of each factor's distribution
percentile = 20
acc_x_pc = np.percentile(abs(data_dropped['acceleration_x']), percentile)
acc_y_pc = np.percentile(abs(data_dropped['acceleration_y']), percentile)
acc_z_pc = np.percentile(abs(data_dropped['acceleration_z']), percentile)
gyro_x_pc = np.percentile(abs(data_dropped['gyro_x']), percentile)
gyro_y_pc = np.percentile(abs(data_dropped['gyro_y']), percentile)
gyro_z_pc = np.percentile(abs(data_dropped['gyro_z']), percentile)
```

```python
acc_x_pc, acc_y_pc, acc_z_pc, gyro_x_pc, gyro_y_pc, gyro_z_pc
```

## Condensed DF with counts of each factor above nth percentile value

```python
# Create empty DF
condensed_counts = pd.DataFrame(0.0, index=np.arange(len(labels)), columns=['bookingID', 'acc_x_count', 'acc_y_count', 'acc_z_count', 'gyro_x_count', 'gyro_y_count', 'gyro_z_count', 'acc_x_max', 'acc_y_max', 'acc_z_max', 'gyro_x_max', 'gyro_y_max', 'gyro_z_max', 'label', 'speed'])
```

```python
# Populate df according to each trip
for i in range(len(labels)):
    condensed_counts.iloc[i,0] = labels.iloc[i,0]
    condensed_counts.iloc[i,13] = labels.iloc[i,1]
    df = data_dropped[data_dropped['bookingID']==labels.iloc[i,0]]
    condensed_counts.iloc[i,14] = max(df['Speed'])
    # Get counts
    condensed_counts.iloc[i,1] = np.sum(abs(df['acceleration_x'])>acc_x_pc)
    condensed_counts.iloc[i,2] = np.sum(abs(df['acceleration_y'])>acc_y_pc)
    condensed_counts.iloc[i,3] = np.sum(abs(df['acceleration_z'])>acc_z_pc)
    condensed_counts.iloc[i,4] = np.sum(abs(df['gyro_x'])>gyro_x_pc)
    condensed_counts.iloc[i,5] = np.sum(abs(df['gyro_y'])>gyro_y_pc)
    condensed_counts.iloc[i,6] = np.sum(abs(df['gyro_z'])>gyro_z_pc)
    # Get max
    condensed_counts.iloc[i,7] = max(abs(df['acceleration_x']))
    condensed_counts.iloc[i,8] = max(abs(df['acceleration_y']))
    condensed_counts.iloc[i,9] = max(abs(df['acceleration_z']))
    condensed_counts.iloc[i,10] = max(abs(df['gyro_x']))
    condensed_counts.iloc[i,11] = max(abs(df['gyro_y']))
    condensed_counts.iloc[i,12] = max(abs(df['gyro_z']))
```

```python
condensed_counts
```

```python
# Simple random forest classifier
RF = RandomForestClassifier(n_estimators=1000, min_samples_leaf=50, oob_score=True, random_state=2, verbose=1)
RF.fit(condensed_counts[['acc_x_count', 'acc_y_count', 'acc_z_count', 'gyro_x_count', 'gyro_y_count', 'gyro_z_count', 'acc_x_max', 'acc_y_max', 'acc_z_max', 'gyro_x_max', 'gyro_y_max', 'gyro_z_max', 'speed']], condensed_counts['label'])
RF.score(condensed_counts[['acc_x_count', 'acc_y_count', 'acc_z_count', 'gyro_x_count', 'gyro_y_count', 'gyro_z_count', 'acc_x_max', 'acc_y_max', 'acc_z_max', 'gyro_x_max', 'gyro_y_max', 'gyro_z_max', 'speed']], condensed_counts['label'])
```

```python
# Save RF model
filename = 'RF.sav'
pickle.dump(RF, open(filename, 'wb'))
```

```python
# Confusion matrix
y_true = condensed_counts['label']
y_pred = RF.predict(condensed_counts[['acc_x_count', 'acc_y_count', 'acc_z_count', 'gyro_x_count', 'gyro_y_count', 'gyro_z_count', 'acc_x_max', 'acc_y_max', 'acc_z_max', 'gyro_x_max', 'gyro_y_max', 'gyro_z_max', 'speed']])
tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
(tn, fp, fn, tp)
```
